```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from fancyplots.loaders import loadmat
from fancyplots.utils import stitch_tm, complex_to_hsv
```

```python
tm = loadmat('./resources/matrice_34_34_exp.mat')['tm_corr']
weights = loadmat('./resources/weights.mat')
```

```python
stitched_tm = stitch_tm(tm, weights, crop_outputs=True, cropped_length=180)
stitched_tm = stitched_tm / np.nanmax(np.abs(stitched_tm))

if np.iscomplexobj(stitched_tm):
    plt.imshow(complex_to_hsv(stitched_tm, rmin=0, rmax=np.max(np.abs(stitched_tm))))
else:
    plt.imshow(stitched_tm, cmap='gray')
```
